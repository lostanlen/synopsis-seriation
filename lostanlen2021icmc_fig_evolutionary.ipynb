```python
import numpy as np
import os

trials = range(100)
trial_strs = [
    "permutations_trial" + str(trial).zfill(3)
    for trial in trials
]
trial_objectives = []

for trial_str in trial_strs:
    trial_dir = os.path.join(
        "/Users/vl238/synopsis_seriation_data/permutations",
        trial_str
    )
    objectives = []
    mats = sorted(os.listdir(trial_dir))
    for mat in mats:
        objectives.append(float(mat[-14:-4]))
    trial_objectives.append(np.array(objectives))
    
plt.figure(figsize=(15, 5))
for objectives in trial_objectives:
    plt.plot(
        100*np.arange(len(objectives)),
        objectives,
        color='k', linewidth=0.5, alpha=0.5
    );
plt.plot(
    100*np.arange(len(objectives)),
    trial_objectives[37], '-',
    color='#FF7F00', linewidth=4, alpha=1.0
);

trials = range(100)
trial_strs = [
    "permutations_v2_trial" + str(trial).zfill(3)
    for trial in trials
]
trial_objectives = []

for trial_str in trial_strs:
    trial_dir = os.path.join(
        "/Users/vl238/synopsis_seriation_data/permutations_v2",
        trial_str
    )
    objectives = []
    mats = sorted(os.listdir(trial_dir))
    for mat in mats:
        objectives.append(float(mat[-14:-4]))
    trial_objectives.append(np.array(objectives))
    
for objectives in trial_objectives:
    plt.plot(
        6100 + 100*np.arange(len(objectives)),
        objectives,
        color='k', linewidth=0.5, alpha=0.2
    );
plt.plot(
    6100 + 100*np.arange(len(trial_objectives[27])),
    trial_objectives[27], '-',
    color='#FF7F00', linewidth=4, alpha=1
);
    
plt.plot([6100, 6100], [0, 2e8], color='blue')
plt.ylim(78e6, 115e6)
plt.xlim(0, 20000)
plt.xlabel("Iterations")
plt.ylabel("Seriation loss")
plt.grid("on", linestyle="--")

plt.savefig("synopsis_seriation_fig_evolutionary.pdf", bbox_inches="tight")
```
